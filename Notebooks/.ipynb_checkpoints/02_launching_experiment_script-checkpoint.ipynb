```python
import os
import inspect 
import sys 
import shutil

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

from PyCausality.TransferEntropy import *

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 
from Modules import utils
```

__1) Importing Input Data-Set__

```python
path = '../Input_Data/data_multi_c1.csv'
df_input = pd.read_csv(path, index_col = 0)
df_input = df_input[ (df_input["prov"] == "Cagliari") | (df_input["prov"] == "Agrigento")]
df_input
```

__2) Launching Experiments__

```python
Exp_steps = ["TE", 
             "TE_SHUFFLE", 
             "ETE"]

name_exp = "base"

for exp_step in Exp_steps: 

    utils.pipeline_exp(df_input,
                       name_exp       = name_exp,
                       Input_features = ['x'],
                       Targets        = ['y'],
                       LAGS_          = [2,3,4,5,6,7,8],
                       EXP_step       = exp_step,
                       N_shuffles     = 10,
                       plot_hist      = False,
                       Study_period   = None,
                       FOLD_save      = "../Output_Experiments/")
```
